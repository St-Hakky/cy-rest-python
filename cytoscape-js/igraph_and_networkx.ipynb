```python
import networkx as nx
import igraph as ig
import json

from py2cytoscape import cytoscapejs as cyjs
import py2cytoscape.util.util_igraph as util_ig 
import py2cytoscape.util.util_networkx as util_nx 
```

```python
print(cyjs.get_layouts())
print(cyjs.get_style_names())
print(cyjs.get_style('default2'))
```

```python
yeast_network = json.load(open('../basic/sample_data/yeast.json')) # Simple yeast PPI network
cyjs.render(yeast_network)
```

```python
g = nx.watts_strogatz_graph(500, 4, 0.35)
pos=nx.graphviz_layout(g,prog='dot')

g_cyjs = util_nx.from_networkx(g, layout=pos, scale=1)
```

```python
cyjs.render(g_cyjs, style='Directed', background='black')
```
